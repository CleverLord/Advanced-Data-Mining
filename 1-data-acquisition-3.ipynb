```python
import pickle
#load tematy_df.pickle and descriptions_df.pickle
with open('tematy_df.pickle', 'rb') as f:
    tematy_df = pickle.load(f)
with open('descriptions_df.pickle', 'rb') as f:
    descriptions_df = pickle.load(f)
```

```python
tematy_df.sample(2)
```

```python
descriptions_df.sample(2)
```

```python
len(tematy_df), len(descriptions_df)
```

```python
#join on url
tematy_i_opisy_df = tematy_df.merge(descriptions_df, on='url')
tematy_i_opisy_df.sample(2)
```

```python
#drop topic, rename description to Temat Pracy
tematy_i_opisy_df = tematy_i_opisy_df.drop('topic', axis=1)
tematy_i_opisy_df = tematy_i_opisy_df.rename(columns={'description': 'Temat Pracy'})
tematy_i_opisy_df.sample(2)
```

```python
#reorder columns so that with current names [Tytuł pracy w języku oryginału	url	Jednostka organizacyjna	Osoba zgłaszająca temat	Kierunki studiów	Typ pracy	Status tematu	Data aktualizacji	Temat Pracy]
#we would have Tytuł pracy w języku oryginału, Temat Pracy, Jednostka organizacyjna, Osoba zgłaszająca temat, Kierunki studiów, Typ pracy, Status tematu, Data aktualizacji, url
tematy_i_opisy_df = tematy_i_opisy_df[['Tytuł pracy w języku oryginału', 'Temat Pracy', 'Jednostka organizacyjna', 'Osoba zgłaszająca temat', 'Kierunki studiów', 'Typ pracy', 'Status tematu', 'Data aktualizacji', 'url']]
#remove pandas print limit per row
import pandas as pd
pd.set_option('display.max_colwidth', None)
tematy_i_opisy_df.sample(2)
```

```python
#save to tematy_i_opisy_df.pickle
with open('tematy_i_opisy_df.pickle', 'wb') as f:
    pickle.dump(tematy_i_opisy_df, f)
```
